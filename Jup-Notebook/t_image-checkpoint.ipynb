```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121, ResNet50, Xception
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Precision, Recall

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
import logging
import pickle
from sklearn.metrics import classification_report, confusion_matrix

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("brain_stroke_cnn.log"),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

# Set random seed for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# Set paths for dataset
DATASET_PATH = 'Brain_Stroke_CT_SCAN_image'
TRAIN_PATH = os.path.join(DATASET_PATH, 'Train')
TEST_PATH = os.path.join(DATASET_PATH, 'Test')
VALID_PATH = os.path.join(DATASET_PATH, 'Validation')

# Set image parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# ==========================================
# 1. Data Loading and Visualization
# ==========================================

# Function to load and prepare datasets
def load_datasets():
    """
    Load and prepare train, validation, and test datasets
    
    Returns:
    --------
    tuple
        train_ds, valid_ds, test_ds
    """

    logger.info("Loading datasets...")
    
    # Check if directories exist
    for path in [TRAIN_PATH, TEST_PATH, VALID_PATH]:
        if not os.path.exists(path):
            raise FileNotFoundError(f"Directory not found: {path}")
    
    # Data augmentation for training
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    
    # Only rescaling for validation and test
    valid_test_datagen = ImageDataGenerator(rescale=1./255)
    
    # Load datasets
    train_ds = train_datagen.flow_from_directory(
        TRAIN_PATH,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        seed=RANDOM_SEED
    )
    
    valid_ds = valid_test_datagen.flow_from_directory(
        VALID_PATH,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        seed=RANDOM_SEED
    )
    
    test_ds = valid_test_datagen.flow_from_directory(
        TEST_PATH,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False,
        seed=RANDOM_SEED
    )
    
    logger.info(f"Classes found: {train_ds.class_indices}")
    logger.info(f"Training samples: {train_ds.samples}")
    logger.info(f"Validation samples: {valid_ds.samples}")
    logger.info(f"Test samples: {test_ds.samples}")
    
    return train_ds, valid_ds, test_ds

# Function to visualize data distribution
def visualize_data_distribution():
    """
    Visualize the distribution of data across classes and sets
    """
    logger.info("Visualizing data distribution...")
    
    # Get class counts for each set
    train_counts = {}
    test_counts = {}
    valid_counts = {}
    
    # Count training images
    for class_name in os.listdir(TRAIN_PATH):
        class_dir = os.path.join(TRAIN_PATH, class_name)
        if os.path.isdir(class_dir):
            train_counts[class_name] = len(os.listdir(class_dir))
    
    # Count test images
    for class_name in os.listdir(TEST_PATH):
        class_dir = os.path.join(TEST_PATH, class_name)
        if os.path.isdir(class_dir):
            test_counts[class_name] = len(os.listdir(class_dir))
    
    # Count validation images
    for class_name in os.listdir(VALID_PATH):
        class_dir = os.path.join(VALID_PATH, class_name)
        if os.path.isdir(class_dir):
            valid_counts[class_name] = len(os.listdir(class_dir))
    
    # Plot distribution
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    
    # Training data
    axs[0].bar(train_counts.keys(), train_counts.values(), color='mediumseagreen')
    axs[0].set_title('Training Data Distribution')
    axs[0].set_ylabel('Number of Images')
    
    # Validation data
    axs[1].bar(valid_counts.keys(), valid_counts.values(), color='royalblue')
    axs[1].set_title('Validation Data Distribution')
    
    # Test data
    axs[2].bar(test_counts.keys(), test_counts.values(), color='tomato')
    axs[2].set_title('Test Data Distribution')
    
    plt.tight_layout()
    plt.show()
    
    # Calculate and print totals
    train_total = sum(train_counts.values())
    valid_total = sum(valid_counts.values())
    test_total = sum(test_counts.values())
    total_images = train_total + valid_total + test_total
    
    print(f"Total images: {total_images}")
    print(f"Training: {train_total} images ({train_total/total_images*100:.1f}%)")
    print(f"Validation: {valid_total} images ({valid_total/total_images*100:.1f}%)")
    print(f"Testing: {test_total} images ({test_total/total_images*100:.1f}%)")
    
    # Create pie chart of overall distribution
    plt.figure(figsize=(10, 6))
    plt.pie(
        [train_total, valid_total, test_total],
        labels=['Training', 'Validation', 'Testing'],
        colors=['mediumseagreen', 'royalblue', 'tomato'],
        autopct='%1.1f%%',
        startangle=90,
        shadow=True
    )
    plt.axis('equal')
    plt.title('Overall Data Distribution')
    plt.show()

# Function to display sample images
def display_sample_images(dataset, num_samples=25, title="Sample Images"):
    """
    Display sample images from a dataset
    
    Parameters:
    -----------
    dataset : tf.keras.preprocessing.image.DirectoryIterator
        Dataset to display images from
    num_samples : int
        Number of images to display
    title : str
        Title for the figure
    """
    logger.info(f"Displaying {num_samples} sample images...")
    
    # Get class names
    class_names = list(dataset.class_indices.keys())
    
    # Get a batch of images and labels
    images, labels = next(dataset)
    
    # Determine grid size
    grid_size = int(np.ceil(np.sqrt(num_samples)))
    
    plt.figure(figsize=(15, 15))
    plt.suptitle(title, fontsize=20)
    
    for i in range(min(num_samples, len(images))):
        plt.subplot(grid_size, grid_size, i+1)
        plt.imshow(images[i])
        
        # Get the class name from one-hot encoded label
        class_idx = np.argmax(labels[i])
        plt.title(class_names[class_idx])
        plt.axis('off')
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()

# Function to visualize sample augmentations
def visualize_augmentations():
    """
    Visualize the effect of data augmentation on sample images
    """
    logger.info("Visualizing data augmentations...")
    
    # Load a single image for demonstration
    for class_name in os.listdir(TRAIN_PATH):
        class_dir = os.path.join(TRAIN_PATH, class_name)
        if os.path.isdir(class_dir):
            image_files = os.listdir(class_dir)
            if image_files:
                img_path = os.path.join(class_dir, image_files[0])
                break
    
    # Read and preprocess the image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    
    # Create a datagen with augmentations
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    
    # Convert to array and reshape
    x = np.expand_dims(img, axis=0)
    
    # Create iterator
    aug_iter = datagen.flow(x, batch_size=1)
    
    # Plot original and augmented images
    plt.figure(figsize=(15, 10))
    plt.subplot(3, 4, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')
    
    for i in range(11):
        aug_img = next(aug_iter)[0].astype('uint8')
        plt.subplot(3, 4, i+2)
        plt.imshow(aug_img)
        plt.title(f'Augmented #{i+1}')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# ==========================================
# 2. Model Building
# ==========================================

# Function to create CNN model with transfer learning
def create_model(base_model_name, input_shape=(224, 224, 3), num_classes=2):
    """
    Create a CNN model using transfer learning
    
    Parameters:
    -----------
    base_model_name : str
        Name of the base model to use ('densenet121', 'resnet50', or 'xception')
    input_shape : tuple
        Input shape for the model
    num_classes : int
        Number of output classes
        
    Returns:
    --------
    Model
        The compiled Keras model
    """
    logger.info(f"Creating {base_model_name} model...")
    
    # Initialize base model
    if base_model_name.lower() == 'densenet121':
        base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name.lower() == 'resnet50':
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name.lower() == 'xception':
        base_model = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    else:
        raise ValueError(f"Unknown base model: {base_model_name}")
    
    # Initially set all layers to be non-trainable
    for layer in base_model.layers:
        layer.trainable = False
    
    # Add custom layers on top
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)
    
    # Final layer with softmax activation for multi-class classification
    predictions = Dense(num_classes, activation='softmax')(x)
    
    # Create the model
    model = Model(inputs=base_model.input, outputs=predictions)
    
    # Compile the model
    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy', Precision(), Recall()]
    )
    
    return model, base_model

# Function to unfreeze layers for fine-tuning
def unfreeze_model(model, base_model, num_layers_to_unfreeze=20):
    """
    Unfreeze the last n layers of the base model for fine-tuning
    
    Parameters:
    -----------
    model : Model
        The full model to fine-tune
    base_model : Model
        The base model component
    num_layers_to_unfreeze : int
        Number of layers to unfreeze from the end
        
    Returns:
    --------
    Model
        The model with unfrozen layers
    """
    logger.info(f"Unfreezing last {num_layers_to_unfreeze} layers for fine-tuning...")
    
    # Unfreeze the last n layers
    for layer in base_model.layers[-num_layers_to_unfreeze:]:
        layer.trainable = True
    
    # Recompile the model with a lower learning rate
    model.compile(
        optimizer=Adam(learning_rate=0.00001),  # Lower learning rate for fine-tuning
        loss='categorical_crossentropy',
        metrics=['accuracy', Precision(), Recall()]
    )
    
    return model

# Function to train the model
def train_model(model, train_ds, valid_ds, model_name, epochs=50, fine_tune_epochs=20):
    """
    Train the model with early stopping and learning rate reduction
    
    Parameters:
    -----------
    model : Model
        The model to train
    train_ds : DirectoryIterator
        Training dataset
    valid_ds : DirectoryIterator
        Validation dataset
    model_name : str
        Name of the model for saving
    epochs : int
        Number of initial training epochs
    fine_tune_epochs : int
        Number of fine-tuning epochs
        
    Returns:
    --------
    tuple
        Trained model and training history
    """
    logger.info(f"Training {model_name} model...")
    
    # Set up callbacks
    callbacks = [
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            verbose=1,
            min_lr=0.00001,
            min_delta=0.0001,
            mode='auto'
        ),
        EarlyStopping(
            monitor='val_loss',
            patience=10,
            verbose=1,
            restore_best_weights=True
        )
    ]
    
    # Initial training with frozen base model
    history = model.fit(
        train_ds,
        epochs=epochs,
        validation_data=valid_ds,
        callbacks=callbacks
    )
    
    # Get base model from the full model
    base_model = model.layers[0] if isinstance(model.layers[0], Model) else None
    
    if base_model:
        # Fine-tune the model by unfreezing some layers
        model = unfreeze_model(model, base_model)
        
        # Continue training with unfrozen layers
        fine_tune_history = model.fit(
            train_ds,
            epochs=fine_tune_epochs,
            validation_data=valid_ds,
            callbacks=callbacks
        )
        
        # Combine histories
        combined_history = {}
        for key in history.history:
            combined_history[key] = history.history[key] + fine_tune_history.history[key]
    else:
        combined_history = history.history
    
    # Save the model
    model.save(f"{model_name}_model.keras")
    
    return model, combined_history

# Function to evaluate the model
def evaluate_model(model, test_ds, model_name):
    """
    Evaluate the model and display results
    
    Parameters:
    -----------
    model : Model
        The trained model to evaluate
    test_ds : DirectoryIterator
        Test dataset
    model_name : str
        Name of the model for display purposes
        
    Returns:
    --------
    dict
        Evaluation metrics
    """
    logger.info(f"Evaluating {model_name} model...")
    
    # Evaluate the model
    results = model.evaluate(test_ds, verbose=1)
    
    # Get metrics
    metrics = dict(zip(model.metrics_names, results))
    logger.info(f"{model_name} Test Results: {metrics}")
    
    # Get true labels and predictions
    y_true = np.array([])
    y_pred = np.array([])
    y_prob = np.array([])
    
    # Reset test dataset
    test_ds.reset()
    
    # Predict in batches
    for i in range(len(test_ds)):
        x, y = test_ds[i]
        batch_pred = model.predict(x, verbose=0)
        batch_true = np.argmax(y, axis=1)
        batch_pred_class = np.argmax(batch_pred, axis=1)
        
        y_true = np.append(y_true, batch_true)
        y_pred = np.append(y_pred, batch_pred_class)
        
        if i == 0:
            y_prob = batch_pred
        else:
            y_prob = np.vstack((y_prob, batch_pred))
    
    # Get class names
    class_names = list(test_ds.class_indices.keys())
    
    # Generate classification report
    report = classification_report(y_true, y_pred, target_names=class_names, digits=4)
    print(f"\nClassification Report for {model_name}:")
    print(report)
    
    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=class_names,
        yticklabels=class_names
    )
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    return metrics

# Function to plot training history
def plot_training_history(history, model_name):
    """
    Plot the training and validation accuracy and loss
    
    Parameters:
    -----------
    history : dict
        Training history
    model_name : str
        Name of the model for display purposes
    """
    logger.info(f"Plotting training history for {model_name}...")
    
    plt.figure(figsize=(12, 5))
    
    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Training Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # If precision and recall are available, plot them
    if 'precision' in history and 'recall' in history:
        plt.figure(figsize=(12, 5))
        
        # Plot precision
        plt.subplot(1, 2, 1)
        plt.plot(history['precision'], label='Training Precision')
        plt.plot(history['val_precision'], label='Validation Precision')
        plt.title(f'{model_name} - Precision')
        plt.xlabel('Epoch')
        plt.ylabel('Precision')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Plot recall
        plt.subplot(1, 2, 2)
        plt.plot(history['recall'], label='Training Recall')
        plt.plot(history['val_recall'], label='Validation Recall')
        plt.title(f'{model_name} - Recall')
        plt.xlabel('Epoch')
        plt.ylabel('Recall')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

# Function to compare models
def compare_models(model_metrics):
    """
    Compare the performance of different models
    
    Parameters:
    -----------
    model_metrics : dict
        Dictionary with model names as keys and their metrics as values
    """
    logger.info("Comparing model performances...")
    
    # Extract metrics for comparison
    models = list(model_metrics.keys())
    accuracy = [model_metrics[model]['accuracy'] * 100 for model in models]
    precision = [model_metrics[model]['precision'] * 100 for model in models]
    recall = [model_metrics[model]['recall'] * 100 for model in models]
    
    # Create a dataframe for easy display
    comparison_df = pd.DataFrame({
        'Model': models,
        'Accuracy (%)': accuracy,
        'Precision (%)': precision,
        'Recall (%)': recall
    })
    
    # Sort by accuracy
    comparison_df = comparison_df.sort_values('Accuracy (%)', ascending=False)
    
    # Display the comparison table
    print("\nModel Performance Comparison:")
    print(comparison_df.to_string(index=False))
    
    # Plot the comparison
    plt.figure(figsize=(12, 6))
    
    x = np.arange(len(models))
    width = 0.25
    
    plt.bar(x - width, accuracy, width, label='Accuracy', color='mediumseagreen')
    plt.bar(x, precision, width, label='Precision', color='royalblue')
    plt.bar(x + width, recall, width, label='Recall', color='tomato')
    
    plt.title('Model Performance Comparison')
    plt.xlabel('Model')
    plt.ylabel('Metric Value (%)')
    plt.xticks(x, models)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return comparison_df

# ==========================================
# 3. Prediction and Visualization
# ==========================================

# Function to generate Grad-CAM visualization
def generate_gradcam(img_path, model, last_conv_layer_name, pred_index=None):
    """
    Generate Grad-CAM visualization for model interpretability
    
    Parameters:
    -----------
    img_path : str
        Path to the image
    model : Model
        Trained model
    last_conv_layer_name : str
        Name of the last convolutional layer in the model
    pred_index : int or None
        Index of the predicted class
        
    Returns:
    --------
    tuple
        Heatmap and superimposed image
    """
    # Load and preprocess image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img_array = np.expand_dims(img.astype('float32') / 255.0, axis=0)
    
    # Create a model that maps the input image to the activations of the last conv layer
    grad_model = tf.keras.models.Model(
        [model.inputs], 
        [model.get_layer(last_conv_layer_name).output, model.output]
    )
    
    # Compute the gradient of the top predicted class with respect to the output of the last conv layer
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    
    # Gradient of the output neuron with respect to the output feature map of the last conv layer
    grads = tape.gradient(class_channel, last_conv_layer_output)
    
    # Vector of mean values of gradients over feature map channels
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Weight the channels by the gradient values
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    # Normalize the heatmap
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    heatmap = heatmap.numpy()
    
    # Resize the heatmap to the original image size
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    
    # Convert heatmap to RGB
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    
    # Superimpose the heatmap on original image
    img_uint8 = np.uint8(255.0 * img / np.max(img))
    superimposed_img = cv2.addWeighted(img_uint8, 0.6, heatmap, 0.4, 0)
    
    return heatmap, superimposed_img

# Function to make predictions on a single image
def predict_image(img_path, models, class_names):
    """
    Make predictions using multiple models and visualize the results
    
    Parameters:
    -----------
    img_path : str
        Path to the image
    models : dict
        Dictionary of trained models
    class_names : list
        List of class names
        
    Returns:
    --------
    dict
        Prediction results
    """
    logger.info(f"Predicting on image: {img_path}")
    
    # Load and preprocess image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img_array = np.expand_dims(img.astype('float32') / 255.0, axis=0)
    
    # Store predictions from all models
    predictions = {}
    ensemble_probs = np.zeros((1, len(class_names)))
    
    # Predict with each model
    for model_name, model in models.items():
        pred = model.predict(img_array, verbose=0)[0]
        pred_class = np.argmax(pred)
        confidence = pred[pred_class]
        
        predictions[model_name] = {
            'probabilities': pred,
            'predicted_class': class_names[pred_class],
            'confidence': confidence
        }
        
        # Add to ensemble
        ensemble_probs += pred
    
    # Calculate ensemble prediction
    ensemble_probs /= len(models)
    ensemble_class = np.argmax(ensemble_probs[0])
    ensemble_confidence = ensemble_probs[0][ensemble_class]
    
    # Create a combined prediction object
    result = {
        'individual_models': predictions,
        'ensemble': {
            'probabilities': ensemble_probs[0],
            'predicted_class': class_names[ensemble_class],
            'confidence': ensemble_confidence
        },
        'original_image': img
    }
    
    # Display the image with predictions
    plt.figure(figsize=(15, 10))
    
    # Original image
    plt.subplot(2, 3, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')
    
    # Individual model predictions
    i = 2
    for model_name, pred in predictions.items():
        plt.subplot(2, 3, i)
        plt.imshow(img)
        plt.title(f"{model_name}\nClass: {pred['predicted_class']}\nConf: {pred['confidence']:.2f}")
        plt.axis('off')
        i += 1
    
    # Ensemble prediction
    plt.subplot(2, 3, i)
    plt.imshow(img)
    plt.title(f"Ensemble Prediction\nClass: {result['ensemble']['predicted_class']}\nConf: {result['ensemble']['confidence']:.2f}")
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Generate and display Grad-CAM visualizations
    plt.figure(figsize=(15, 10))
    
    # Original image
    plt.subplot(2, 2, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')
    
    # GradCAM for each model
    i = 2
    for model_name, model in models.items():
        # Get the last convolutional layer name
        last_conv_layer = None
        for layer in reversed(model.layers):
            if 'conv' in layer.name:
                last_conv_layer = layer.name
                break
        
        if last_conv_layer:
            # Generate GradCAM
            _, superimposed = generate_gradcam(
                img_path,
                model,
                last_conv_layer,
                pred_index=ensemble_class
            )
            
            plt.subplot(2, 2, i)
            plt.imshow(superimposed)
            plt.title(f"{model_name} GradCAM")
            plt.axis('off')
            i += 1
    
    plt.tight_layout()
    plt.show()
    
    return result

# ==========================================
# 4. Main Execution
# ==========================================

def main():
    """
    Main function to execute the entire pipeline
    """
    logger.info("Starting brain stroke classification with transfer learning")
    
    # Load datasets
    train_ds, valid_ds, test_ds = load_datasets()
    
    # Data visualization
    visualize_data_distribution()
    display_sample_images(train_ds, title="Training Images")
    visualize_augmentations()
    
    # Model training and evaluation
    models = {}
    histories = {}
    metrics = {}
    model_names = ['DenseNet121', 'ResNet50', 'Xception']
    
    for model_name in model_names:
        # Create model
        model, base_model = create_model(
            model_name.lower(),
            input_shape=(*IMG_SIZE, 3),
            num_classes=len(train_ds.class_indices)
        )
        
        # Train model
        trained_model, history = train_model(
            model,
            train_ds,
            valid_ds,
            model_name,
            epochs=50,
            fine_tune_epochs=20
        )
        
        # Plot training history
        plot_training_history(history, model_name)
        
        # Evaluate model
        model_metric = evaluate_model(trained_model, test_ds, model_name)
        
        # Save results
        models[model_name] = trained_model
        histories[model_name] = history
        metrics[model_name] = model_metric
    
    # Compare models
    comparison_df = compare_models(metrics)
    
    # Make predictions on sample images
    for class_name in os.listdir(TEST_PATH):
        class_dir = os.path.join(TEST_PATH, class_name)
        if os.path.isdir(class_dir):
            image_files = os.listdir(class_dir)
            if image_files:
                sample_path = os.path.join(class_dir, image_files[0])
                predict_image(sample_path, models, list(train_ds.class_indices.keys()))
                break
    
    logger.info("Brain stroke classification pipeline completed successfully")

if __name__ == "__main__":
    main()
```
